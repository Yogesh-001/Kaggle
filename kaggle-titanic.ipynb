```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('../input/titanic-csv/train.csv')
```

```python
data.head()
```

```python
data1=pd.read_csv('../input/titanic-csv/test.csv')
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked']=df['Embarked'].dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
S=LabelEncoder()
df['Sex']=S.fit_transform(df['Sex'])
```

```python
df.dtypes
```

```python
X=df.drop(['PassengerId','Survived','Name','Ticket','Fare','Cabin','Embarked'],axis=1)
Y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
print(y_test[:10])
```

```python
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
